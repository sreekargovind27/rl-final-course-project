```python
import random
import gymnasium as gym
from gymnasium import spaces
```

```python
file_path = "diabetic_data.csv"
df = pd.read_csv(file_path)
```

```python
filtered_df = df[df['insulin'] != 'No']
```

```python

glucose_map = {
    'None': np.nan,
    'Norm': 100,
    '>200': 250,
    '>300': 350
}
filtered_df['glucose_level'] = filtered_df['max_glu_serum'].map(glucose_map)
```

```python
insulin_map = {
    'Down': 0,
    'Steady': 1,
    'Up': 2
}
filtered_df['insulin_action'] = filtered_df['insulin'].map(insulin_map)


filtered_df['reward'] = filtered_df['readmitted'].apply(
    lambda x: -1 if x in ['<30', '>30'] else 1
)
```

```python
rl_data = filtered_df[['glucose_level', 'insulin_action', 'reward']].dropna()
```

```python
rl_data
```

```python

class RealPatientDosageEnv(gym.Env):
    def __init__(self, data: pd.DataFrame):
        super(RealPatientDosageEnv, self).__init__()
        self.data = data.reset_index(drop=True)
        self.n = len(data)
        self.current_idx = 0
        self.observation_space = spaces.Box(low=0.0, high=1.0, shape=(1,), dtype=np.float32)
        self.action_space = spaces.Discrete(3)

    def reset(self):
        self.current_idx = random.randint(0, self.n - 2)
        glucose = self.data.loc[self.current_idx, 'glucose_level']
        self.state = np.array([glucose / 400], dtype=np.float32)
        return self.state

    def step(self, action):
        true_action = self.data.loc[self.current_idx, 'insulin_action']
        reward = self.data.loc[self.current_idx, 'reward']
        if action != true_action:
            reward -= 0.1

        self.current_idx += 1
        done = self.current_idx >= self.n

        if not done:
            glucose = self.data.loc[self.current_idx, 'glucose_level']
            self.state = np.array([glucose / 400], dtype=np.float32)
        else:
            self.state = np.zeros((1,), dtype=np.float32)

        return self.state, reward, done, {}
```

```python
class QLearningAgent:
    def __init__(self, state_bins=10, n_actions=3, alpha=0.1, gamma=0.9, epsilon=0.1):
        self.state_bins = state_bins
        self.n_actions = n_actions
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.q_table = np.zeros((self.state_bins, self.n_actions))

    def discretize_state(self, state):
        return int(np.clip(state[0] * self.state_bins, 0, self.state_bins - 1))

    def select_action(self, state):
        s = self.discretize_state(state)
        if random.random() < self.epsilon:
            return random.randint(0, self.n_actions - 1)
        return np.argmax(self.q_table[s])

    def update(self, state, action, reward, next_state):
        s = self.discretize_state(state)
        s_next = self.discretize_state(next_state)
        best_next = np.max(self.q_table[s_next])
        td_target = reward + self.gamma * best_next
        td_error = td_target - self.q_table[s, action]
        self.q_table[s, action] += self.alpha * td_error
```

```python
def train_q_agent(env, agent, episodes=5000):
    for ep in range(episodes):
        state = env.reset()
        done = False
        while not done:
            action = agent.select_action(state)
            next_state, reward, done, _ = env.step(action)
            agent.update(state, action, reward, next_state)
            state = next_state
    return agent.q_table

env = RealPatientDosageEnv(rl_data)
agent = QLearningAgent(state_bins=20)
q_table = train_q_agent(env, agent, episodes=5000)
```

```python

q_table_df = pd.DataFrame(q_table, columns=['Action: Decrease', 'Action: Steady', 'Action: Increase'])
q_table_df.index.name = 'Glucose Bin (Discretized)'


```
